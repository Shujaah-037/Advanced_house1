## Solving or reducing error that Input contains NaN, infinity or a value too large for dtype('float64')

```python
import pandas as pd
import numpy as np
df = X_train
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
import pandas as pd
import numpy as np
df1 = y_train
def clean_dataset(df1):
    assert isinstance(df1, pd.DataFrame), "df1 needs to be a pd.DataFrame"
    df1.dropna(inplace=True)
    indices_to_keep = ~df1.isin([np.nan, np.inf, -np.inf]).any(1)
    return df1[indices_to_keep].astype(np.float64)
```
